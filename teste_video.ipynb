```python
from ultralytics import YOLO
import cv2

# Carregar o modelo
model = YOLO('C:\\Users\\pchau\\Documents\\TCC\\yolov8-env\\runs\\detect\\train_11_no_prep_no_HP\\weights\\best.pt')  # Substitua pelo caminho do seu modelo

# Abrir o vdeo
cap = cv2.VideoCapture('C:\\Users\\pchau\\Documents\\TCC\\videos\\video3.mp4')

# Definir sada do vdeo
output_path = "C:\\Users\\pchau\\Documents\\TCC\\videos\\output_video.mp4"
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
fps = int(cap.get(cv2.CAP_PROP_FPS))
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model.predict(frame)  # Detectar objetos no frame
    result = results[0]
    boxes = result.boxes

    # Verifique se existem caixas detectadas
    if len(boxes) > 0:
        for box in boxes:  # Para cada caixa de deteco (primeiro item da lista)
            x1, y1, x2, y2 = box.xyxy[0]  # Coordenadas no formato (x1, y1, x2, y2)
            confidence = box.conf[0]  # Confiana da deteco
            class_id = box.cls[0]  # ID da classe da deteco

            # Desenha a caixa no frame
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
            
            # Coloca o texto sobre a caixa com a classe e a confiana
            label = f"{result.names[int(class_id)]} {confidence:.2f}"
            cv2.putText(frame, label, (int(x1), int(y1) - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)


    # Mostrar o resultado
    out.write(frame)
    #cv2.imshow('frame', frame)
    #if cv2.waitKey(1) & 0xFF == ord('q'):
    #    break

# Fechar o vdeo
cap.release()
out.release()
#cv2.destroyAllWindows()
```
